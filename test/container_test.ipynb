```python
import nextpy.interfaces.jupyter as xtj
from pathlib import Path

@xtj.component
def containers():
    xtj.style(Path('hello.css'))
    with xtj.AppLayout():
        with xtj.sidebar():
            with xtj.box(style='height: 30px;'):
                xtj.markdown(label='Hey')

        # BOX 
        # UISNG WITH 
        with xtj.box(class_name='hell'):
            xtj.text('Box text'),xtj.input(label='Box input')
        #  USING CHILDREN
        xtj.box(
            children = [xtj.text('Box text'),xtj.input(label='Box input')], 
            style='border:2px solid black; padding:2px;'
        )
    
    
containers()
```
